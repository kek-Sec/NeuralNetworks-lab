```python
import numpy as np
from pandas import *
import matplotlib.pyplot as plt
import neurolab as nl
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
X = wine.data
T = wine.target
```

```python
df = DataFrame(wine.data,columns=wine.feature_names)
df.head()
```

### The targets have the values [0,1,2] 
### to train our network we need to transform to bit format

```python
T = np.empty([178,3])
T[:,0] = 0
T[0:131,1] = 0
T[130:,1] = 1
T[0:59,2] = 0
T[59:130,2] = 1
T[130:,2] = 0
```

## Data Split
### We will devide the data into 2 sets.The first set for training(70%) the second for testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,T,test_size=0.30)
```

## Preprocessing the data
### We will use the MinMaxScaler to transform our data to a range [0,1]

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_train = min_max_scaler.fit_transform(X_train)
X_test= min_max_scaler.fit_transform(X_test)
```

## Training

```python
net = nl.net.newff(nl.tool.minmax(X),[5,3])

net.layers[-1].transf = nl.trans.LogSig()
net.layers[0].transf = nl.trans.LogSig()
net.layers[1].transf = nl.trans.LogSig()

#train
error = net.trainf = nl.train.train_rprop(net,X,T,epochs=10000,show=10,lr=0.005,goal=0.00005)

plt.plot(error)
plt.xlabel("Epoch number")
plt.ylabel("Train error")
plt.grid()
plt.show()
```

## Network performance

```python
error[-1] #error from last epoch

#simulate
out = net.sim(X_test)
out = np.around(out)

#accuracy
correct=(out==y_test).all(axis=1)
acc=(np.sum(correct)/len(out))*100
```

```python
acc
```

## Compare the results

```python
np.array_equal(out,y_test)
```

```python
plt.figure(1)
plt.subplot(2,1,1)
plt.plot(out,'ro')
plt.title("out")
plt.grid(True)
plt.subplot(2,1,2)
plt.plot(y_test,'ro')
plt.title("y_test")
plt.grid(True)

plt.tight_layout()
plt.show()
```
