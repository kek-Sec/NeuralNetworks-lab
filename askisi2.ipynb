```python
import neurolab as nl
import matplotlib.pyplot as plt
import numpy as np
```

```python
#setup input
inp = np.array([[0.22, 0.34], [0.21, 0.37], [0.25, 0.31],[0.76, 0.19],[0.84, 0.14]]);
```

```python
#labels
labels = np.array([0,0,0,1,1])
#reshape based on input
labels = labels.reshape(inp.shape[0],1) 
```

```python
# Plot input
plt.plot(inp)
plt.xlabel('width')
plt.ylabel('height')
plt.grid()
plt.show();
```

```python
# Create net with 2 inputs and 1 neuron
net = nl.net.newp(nl.tool.minmax(inp), 1)

# train with delta rule (default)
error = net.train(inp, labels,epochs=100, show=2, lr=0.3)

# Plot results
plt.plot(error)
plt.xlabel('Epoch number')
plt.ylabel('Train error')
plt.grid()
plt.show();
```

```python
test_input = np.array([[0.83 , 0.14],[0.27 , 0.42]])
```

```python
net.sim(test_input)
```
