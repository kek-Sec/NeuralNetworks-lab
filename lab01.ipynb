# Hard limit function

```python
import numpy as np
import matplotlib.pyplot as plt
import neurolab as nl
```

```python
f = nl.trans.HardLim()
x = np.arange(-5,6,0.1)
plt.plot(x,f(x));
```

# Linear Transfer Function

```python
f = nl.trans.PureLin()
x = np.arange(-5, 6, 0.1)
plt.plot(x,f(x));
```

# Logarithmic Sigmoid transfer function 

```python
f = nl.trans.LogSig()
x = np.arange(-5, 6, 0.1)
plt.plot(x,f(x));
```

# Hyperbolic tangent sigmoid transfer function

```python
f = nl.trans.TanSig()
x = np.arange(-5, 6, 0.1)
plt.plot(x,f(x));
```

# Saturating linear transfer function

```python
f = nl.trans.SatLin()
x = np.arange(-5, 6, 0.1)
plt.plot(x,f(x));
```

# Soft max transfer function

```python
f = nl.trans.SoftMax()
x = np.array([4, 5, 6])
plt.plot(x,f(x));
print(f(x))
```

# Python implementation (1st approach)

```python
p = np.array([2,4,6,8]) # a random input vector
w = np.array([0.1,0.2,0.3,0.4]) # a random weight vector
b = 0.5 # the bias
s = 0
f = np.empty([4])
for i in range(4):
    f[i] = p[i]*w[i]
    s = s+f[i]
n = s+b # the output after summation
f = nl.trans.LogSig()
a = f(n) # the final output
```
