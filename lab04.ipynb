# Example 2

```python
import matplotlib.pyplot as plt
import numpy as nmp
```

```python
X = nmp.array([[-0.5,-0.5,0.3,-0.1],[-0.5,0.5,-0.5,1.0]]);
T = nmp.array([[1,1],[1,1],[0,0],[0,0]]);
plt.scatter(X[0],X[1],c=T[:,0]);
```

```python
X = X.T #transpose
```

```python
net = nl.net.newff(nl.tool.minmax(X),[5,2])
```

```python
net.layers[-1].transf = nl.trans.LogSig() #output layer
```

```python
net.layers[0].transf = nl.trans.LogSig() #hidden layer
```

```python
net.layers[1].transf = nl.trans.LogSig() #hidden layer
```

```python
#train the network
error = net.trainf = nl.train.train_rprop(net,X,T,epochs=100,show=12,lr=0.05,goal=0.001)
```

```python
error = net.trainf = nl.train.train_gd(net,X,T,epochs=100,show=12,lr=0.05,goal=0.001)
```

```python
#plot
plt.plot(error)
plt.xlabel("epoch num")
plt.ylabel("Train error")
plt.grid()
plt.show()
```
