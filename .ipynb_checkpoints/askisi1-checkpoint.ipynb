### Modify the code of the second Python implementation in order to give us results when we have the ‘data’ from Iris Dataset as an input.

```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
import neurolab as nl
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
X
```

```python
y
```

```python
input_vector = y
weight_vector = X
```

```python
bias = 0.5
```

```python
characteristic_values = 4 #from iv
neurons = 4
```

```python
output = np.empty([1,neurons])
n = np.empty([characteristic_values,neurons])
```

```python
for i in range(characteristic_values):
    for j in range(neurons):
        n[i,j] = weight_vector[i,j]*input_vector[j]
        output[0,i] = output[0,i]+n[i,j]
    output[0,i] = output[0,i]+bias
    f = nl.trans.LogSig()
    output[0,i] = f(output[0,i]) 
```
